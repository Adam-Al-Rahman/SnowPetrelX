```python
from rich import print

import pandas as pd
import numpy as np

file = "../datasets/dataset.csv"

# SNOW PETREL FEATURE
headers = ["behavior", "image_id", "image_file", "head_x", "head_y", "beak_base_x", "beak_base_y", 
                  "beak_tip_x", "beak_tip_y", "neck_x", "neck_y", "body1_x", "body1_y", 
                  "body2_x", "body2_y", "tail_base_x", "tail_base_y"]
```

```python

SKIP_ROWS = 3
SKIP_COLS = 0
usecols = range(SKIP_COLS, len(headers))

dataset = pd.read_csv(file, skiprows=SKIP_ROWS, usecols=usecols, names=headers)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image using PIL (Pillow)
image_path = '../datasets/preening/p_001/81-20151231002642-00.jpg'
image = Image.open(image_path)

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Turn off axes
plt.show()

# Get image details
width, height = image.size
print(f"Image Size (Width x Height): {width} x {height}")

# Convert image to NumPy array to find shape
image_array = np.array(image)
print(f"Image Shape: {image_array.shape}")  # Output will be (Height, Width, Channels)
```
